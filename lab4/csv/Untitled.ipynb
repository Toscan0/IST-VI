```python
tmp = "/home/pedro/Desktop/VI/VI_Project/lab4/csv/ethnicityWinnerVsNominee.csv"
data = pd.read_csv(tmp)
```

```python
dataframe = pd.DataFrame(data)
```

```python
new_dict = { "year": [] , "race_ethnicity": [], "milestone": [], "count":[]}
```

```python
dataframe.head()
```

```python
dataframe = dataframe.groupby("Year").sum()
```

```python
import numpy as np

arr = []

x = 1928

while x <= 2015:
    arr.append(x)
    x+=1
arr = np.array(arr)

dataframe["Year"] = arr
```

```python
dataframe.head(20)
```

```python
for index,row in dataframe.iterrows():
    if str(row["count_white"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("white")
        new_dict["count"].append(row["count_white"])
        
    if str(row["count_black"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("black")
        new_dict["count"].append(row["count_black"])
   
    if str(row["count_asian"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("asian")
        new_dict["count"].append(row["count_asian"])
        
    if str(row["count_hispa"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("hispan")
        new_dict["count"].append(row["count_hispa"])
        
    if str(row["count_arab"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("arab")
        new_dict["count"].append(row["count_arab"])
        
    if str(row["count_indig"]) != str(0):
        new_dict["year"].append(row["Year"])
        new_dict["race_ethnicity"].append("indig")
        new_dict["count"].append(row["count_indig"])
        
        
        
    new_dict["milestone"] = ""
```

```python
new_df = pd.DataFrame(new_dict)
```

```python
new_dict = {  "year" : [], "count" : [], "milestone" : [], "race_ethnicity" : [] , "sentiment" : []}


for index,row in tmp.iterrows():
    if str(row["milestone.1"]) == str("nan"):
        new_dict["milestone"].append(-1)
        new_dict["sentiment"].append(-1)
    else:
        new_dict["milestone"].append(row["milestone.1"])
        new_dict["sentiment"].append(row["sentiment"])
        
    new_dict["count"].append(row["count"])
    new_dict["race_ethnicity"].append(row["race_ethnicity"])
    new_dict["year"].append(row["year"])
    
```
