```python
def read_files(file):
    df = pd.DataFrame(pd.read_csv(str(mypath) + "/" + str(file)))
    
    if(str(file) == "best_film_editing.csv"):
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Film Editing")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())

    if(str(file) == "best_writing_original_screenplay.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Original Screenplay")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Reference"].strip())
            new_dict["status"].append(row["Status"].strip())

    if(str(file) == "best_actor_in_a_leading_role.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Actor in a Leading Role")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass

    if(str(file) == "best_writing_adapted_screenplay.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Adapted Screenplay")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
        
    if(str(file) == "best_actress_in_a_leading_role.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Actress in a Leading Role")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
    
    if(str(file) == "best_picture.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Picture")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
        
    if(str(file) == "best_music_original_song.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Original Song Score or Adaptation")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
      
    if(str(file) == "best_short_film_animated.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Animated Short Film")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
       
    if(str(file) == "best_music_original_song_score_or_adaptation.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Original Song Score or Adaptation")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass

    if(str(file) == "best_short_film_live_action.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Live Action Short Film")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
    
    if(str(file) == "best_makeup_&_hairstyling.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Makeup and Hairstyling")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
    
    if(str(file) == "best_actor_in_a_supporting_role.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Actor in a Supporting Role")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
        
    if(str(file) == "best_actress_in_a_supporting_role.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Actress in a Supporting Role")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            pass
        
    if(str(file) == "best_actress_in_a_supporting_role.csv"):
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Actress in a Supporting Role")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            
    if(str(file) == "best_animated_feature.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Animated Feature")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())

    if(str(file) == "best_production_design.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Production Design")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())

    if(str(file) == "best_cinematography.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Cinematography")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            

    if(str(file) == "best_director.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Director")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            
    if(str(file) == "best_sound_mixing.csv"):
        df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        
        for index, row in df.iterrows():
            new_dict["year"].append(row["Year"].strip())
            new_dict["film"].append(row["Film"].strip())
            new_dict["award"].append("Best Sound Mixing")
            new_dict["person"].append(row['Name'].strip())
            new_dict["orientation"].append(row["Demographic"].strip())
            new_dict["status"].append(row["Status"].strip())
            
            
    return df

            
tmp = read_files("best_sound_mixing.csv")
tmp = read_files("best_director.csv")
tmp = read_files("best_cinematography.csv")
tmp = read_files("best_film_editing.csv")
tmp = read_files("best_writing_original_screenplay.csv")
tmp = read_files("best_actor_in_a_leading_role.csv")
tmp = read_files("best_writing_adapted_screenplay.csv")
tmp = read_files("best_actress_in_a_leading_role.csv")
tmp = read_files("best_picture.csv")  
tmp = read_files("best_music_original_song.csv")
tmp = read_files("best_short_film_animated.csv")
tmp = read_files("best_documentary_short_subject.csv")
tmp = read_files("best_short_film_live_action.csv")
tmp = read_files("best_makeup_&_hairstyling.csv")
tmp = read_files("best_actor_in_a_supporting_role.csv")
tmp = read_files("best_actress_in_a_supporting_role.csv")
tmp = read_files("best_animated_feature.csv")


#tmp = read_files("best_writing_original_story.csv") #nao tem demographic
#tmp = read_files("best_music_original_score.csv")
#tmp = read_files("best_production_design.csv")      #nao tem demographic
#tmp = read_files("best_writing_original_story.csv") #nao tem demographic
#tmp = read_files("best_music_original_score.csv")   #nao tem demogrpahic
#tmp = read_files("best_production_design.csv")
#tmp = read_files("best_documentary_feature.csv")
#tmp = read_files("best_documentary_feature.csv")                      #nao tem demographic
#tmp = read_files("best_music_original_song_score_or_adaptation.csv")  #nao tem demographic
#tmp = read_files("best_costume_design.csv")                           #nao tem demographic
```
